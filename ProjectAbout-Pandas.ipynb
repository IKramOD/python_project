```python
# Data Cleaning 
df.isnull().sum()

# Dropping all missing values
df.dropna(inplace=True)
```

```python
df[df.duplicated()]

# Deleting all duplicates 
df.drop_duplicates(inplace=True)
```

```python
df.dtypes

df['id'] = df['id'].astype(object)

df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

```python
# Data Analysis
# Price Distribution
df['price']
```

```python
sns.histplot(data= df, x='price')
```

```python
# Identifying outliers
df1 = df[df['price'] <1500]
sns.boxplot(data= df1, x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data = df1, x='price', bins=100)
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,5))
sns.histplot(data = df1, x='availability_365')
plt.ylabel('Frequency')
plt.title('availability_365 Distribution')
plt.show()
```
